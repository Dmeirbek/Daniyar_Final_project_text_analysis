**Research Project: Narratives of the Latin Script Transition in Kazakhstan**

**Background:**
The transition of Kazakhstan from Cyrillic to a Latin-based script is a nuanced narrative intricately intertwined with its historical background, identity, and the impact of globalization. The transition process spans nearly a century, marked with twists and turns that reflect the country's dynamic relationship with its past and its aspirations for the future. This story is one of great significance, and its narration is a testament to the evolution of Kazakhstan as a nation.

Before the Soviet era, the Kazakh language utilized a Latin-based script, briefly introduced between 1929 and 1938. This was, however, replaced by Cyrillic in 1940, aligning Kazakhstan with the broader Soviet linguistic landscape. In the aftermath of independence in 1991, the concept of a Latin script resurfaced, driven by aspirations to distance itself from the Soviet legacy and strengthen connections with the rest of the world, particularly the Western and Turkic world.

2017, President Nazarbayev issued a decree that mandated the gradual transition from Cyrillic to Latin by 2025. The deadline was later revised to 2031. As part of this transition, a new 31-letter alphabet was designed, which included diacritics for unique Kazakh sounds. Pilot projects were initiated in education and media to test the effectiveness of the new script. The government also began publishing official documents in both scripts. The transition to the Latin script, as mandated by the decree, is expected to be completed by 2031.

Preparing vast educational materials, training educators, and navigating public opinion divided between advocates and sceptics pose significant challenges for transition. The concerns regarding the cost, potential disruption to literacy, and the impact on minority languages cannot be overlooked.

The transition to Latin script has garnered a mixed reaction both inside the country and across the globe. Although no study has been done to discover local population opinion on the Latin script transition, the different perspectives among Kazakhstani society highlight the complexities inherent in promoting linguistic diversity while navigating diverse communities' political and cultural sensitivities.

Advocates of the proposal assert that it will positively impact the Kazakh identity and facilitate digital access and integration with the globalized world, which is dominated by the Latin script. They believe this move will elevate the prestige of the Kazakh language and create new economic and cultural opportunities.

On the other hand, critics are concerned about the potential consequences of the transition, particularly concerning the possible alienation of older generations, marginalization of Russian speakers, and disruption of national unity. Some have also questioned the transition's economic benefits and suggested that resources may be better allocated to other priorities.

The Kazakh government's transition from Cyrillic to Latin script has generated widespread interest worldwide too. Government officials hold divergent views on the matter. They believe the Western world views this development positively as a step towards establishing closer ties between nations. Correspondingly, it is perceived that Western world tends to view the transition as a progressive move away from the Soviet legacy and embracing democratic values and economic development.

It is perceived in Kazakhstani society that Russia has different interpretations regarding the shift from the Cyrillic to the Latin alphabet in Kazakhstan. Government officials in Kazakhstan generally think that Moscow sees it as a calculated distancing from their cultural influence and a repudiation of their shared Soviet past. Kazakhstan's historical ties and significant Russian minority add geopolitical complexity to the script change. Moscow may view this move with suspicion and seek ways to exert pressure or slow down the process, fearing a decline in their sway. This development is not only about the alphabet but also about changing alliances and national identity in the post-Soviet landscape.

The journey of Kazakhstan towards adopting the Latin script is an ongoing process that requires careful consideration of practicalities, concerns, and long-term impacts. The next decade will be vital in determining the success of this ambitious linguistic transformation. The transformation could either unify the nation or lead to further divisions. However, Kazakhstan's script is currently in flux, reflecting a country in transition, grappling with its past, and striving to find its place in the world.

**Research Question and Hypothesis:**
Given the intricacies surrounding the transition to the Latin script in Kazakhstan, which carries significant domestic and international political implications,  it was crucial to comprehensively examine public opinion concerning this transition domestically in Kazakhstan, Russia, and the Western sphere. An analysis of the prevalent discourse in widely accessed online news sources offers a relevant method for discerning diverse perspectives on the matter. Such an inquiry would facilitate a nuanced understanding of the prevailing circumstances, thereby contributing to a smoother transition process considering a spectrum of viewpoints.

The initial research question that guided this inquiry was: How is the shift of Kazakhstan to the Latin script represented across various online news sources, encompassing Kazakh, Russian, and Western media?
```python
# function to filter and tokenize words, remove capitalization, numbers and punctuation
def filter_and_tokenize(text):
    text_tokens = [t.lower() for t in text.split() if t.isalpha()]
    return text_tokens

# applying the function to each row of the 'text1' column
data_df['text1_tokens'] = data_df['text'].apply(lambda x: filter_and_tokenize(x))

#creating a new column with the every speech without stopwords
stops = set(stopwords.words('english'))
data_df['text1_stops'] = data_df['text1_tokens'].apply(lambda tokens: [t for t in tokens if t not in stops])

#Lemmatizing
wordnet_lemmatizer = WordNetLemmatizer()

def get_wordnet_pos(word):
    """Map POS tag to first character lemmatize() accepts"""
    tag = nltk.pos_tag([word])[0][1][0].upper()
    tag_dict = {"J": wordnet.ADJ,
                "N": wordnet.NOUN,
                "V": wordnet.VERB,
                "R": wordnet.ADV}

    return tag_dict.get(tag, wordnet.NOUN)

data_df['text1_clean'] = data_df['text1_stops'].apply(lambda tokens: [wordnet_lemmatizer.lemmatize(t, get_wordnet_pos(t)) for t in tokens])
```

```python
import pandas as pd # now i start Keyword Analysis for Sentiment Categories
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize
from collections import Counter
```

```python
from nltk.corpus import wordnet

positive_keywords = ['development', 'Identity', 'culture', 'positive', 'Change', 'Shift', 'identity', 'cultural', 'modernization', 'positive','impact', 'progress', 'prosperity']
negative_keywords = ['political','geopolitical', 'conflict', 'tension', 'struggle', 'opposition', 'dispute', 'challenges', 'conflict']
neutral_keywords = ['transition', 'change', 'policy shift']

def find_similar_words(word):
    similar_words = set()
    synsets = wordnet.synsets(word)
    for synset in synsets:
        similar_words.update([lemma.name() for lemma in synset.lemmas()])
    return similar_words


# create new lists to store the extended terms
extended_positive_words = positive_keywords.copy()
extended_negative_words = negative_keywords.copy()
extended_neutral_words = neutral_keywords.copy()


# find similar words for each term and update the lists
for term in positive_keywords:
    similar_words = find_similar_words(term)
    extended_positive_words.extend(similar_words)

for term in negative_keywords:
    similar_words = find_similar_words(term)
    extended_negative_words.extend(similar_words)

for term in neutral_keywords:
    similar_words = find_similar_words(term)
    extended_neutral_words.extend(similar_words)

# apply the function to each row in the "text1_clean" column and print the results
for index, row in data_df.iterrows():
    for term in extended_positive_words:
        similar_words = find_similar_words(term)
        print(f"Similar words for {term} in row {index + 1}: {similar_words}")

    for term in  extended_negative_words:
        similar_words = find_similar_words(term)
        print(f"Similar words for {term} in row {index + 1}: {similar_words}")

    for term in extended_neutral_words:
        similar_words = find_similar_words(term)
        print(f"Similar words for {term} in row {index + 1}: {similar_words}")

```

```python
import matplotlib.pyplot as plt

# function to count the number of occurrences of terms in a text
def count_term_occurrences(text, terms):
    return sum(1 for term in terms if term in text)

# create lists to store the counts for each speech
positive_counts = []
negative_counts = []
neutral_counts = []

# Iterate through each row and count occurrences
for index, row in data_df.iterrows():
    text = row['text1_clean']

    extended_positive_words = positive_keywords.copy()
    extended_negative_words = negative_keywords.copy()
    extended_neutral_words = neutral_keywords.copy()

    # count occurrences of positive terms
    positive_count = count_term_occurrences(text, extended_positive_words)
    positive_counts.append(positive_count)

    # count occurrences of negative terms
    negative_count = count_term_occurrences(text, extended_negative_words)
    negative_counts.append(negative_count)

    # count occurrences of neutral terms
    neutral_count = count_term_occurrences(text, extended_neutral_words)
    neutral_counts.append(neutral_count)

# plot the results for positive terms
plt.figure(figsize=(10, 5))
plt.bar(range(len(data_df)), positive_counts, label='Positive Words', alpha=0.7, color='blue')
plt.xlabel('Article')
plt.ylabel('Number of Positive Occurrences')
plt.title('Occurrences of Positive Terms in Articles')
plt.show()

# plot the results for negative terms
plt.figure(figsize=(10, 5))
plt.bar(range(len(data_df)), negative_counts, label='Negative Words', alpha=0.7, color='blue')
plt.xlabel('Article')
plt.ylabel('Number of Negative Occurrences')
plt.title('Occurrences of Negative Terms in Articles')
plt.show()

# plot the results for neutral terms
plt.figure(figsize=(10, 5))
plt.bar(range(len(data_df)), neutral_counts, label='Neutral Words', alpha=0.7, color='blue')
plt.xlabel('Article')
plt.ylabel('Number of Neutral Occurrences')
plt.title('Occurrences of Neutral Terms in Articles')
plt.show()


```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

def generate_wordcloud_for_common_words(data_df, extended_positive_words):
    # create an empty list to store text with common words
    common_texts = []

    # Iterate through each row and check for common words
    for index, row in data_df.iterrows():
        text_clean = row['text1_clean']

        # find common words
        common_words = [word for word in text_clean if word in extended_positive_words]

        # join the common words to form a text
        common_text = ' '.join(common_words)
        common_texts.append(common_text)

    # join all the common texts
    all_common_text = ' '.join(common_texts)

    # generate and display WordCloud for common words
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(all_common_text)

    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title('Word Cloud for Positive Words')
    plt.show()



# call the function
generate_wordcloud_for_common_words(data_df, extended_positive_words)
```
